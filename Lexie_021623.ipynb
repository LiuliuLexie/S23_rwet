```python
text = open("story_of_your_life.txt").read()
```

```python
text
```

```python
para = text.split("\n")
lines = text.split(". ")
```

```python
import random
random.choice(para)
```

```python
random.choice(lines)
```

```python
x = random.randrange(20, 50, 2)
y = random.randrange(50, 100, 2)

short_lines = [item for item in lines if len(item) < x]
long_lines = [item for item in lines if x < len(item) < y]

stanza_count = 5
randomline = [short_lines, long_lines]
for i in range(stanza_count):
    print()
    print(random.choice(short_lines))
    print("  " + random.choice(random.choice(randomline)) + "\n")
    print("    " + random.choice(random.choice(randomline)) + "\n")
    print("      " + random.choice(random.choice(randomline)) + "\n")
```
