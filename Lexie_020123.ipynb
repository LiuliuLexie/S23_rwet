## Diary Generator

```python
date = [
    "Monday",
    "Tuesday",
    "Wednesday",
    "Thursday",
    "Friday",
    "Saturday",
    "Sunday"]
```

```python
weather = [
    "sunny ☀️",
    "snowy ❄️",
    "cloudy ☁️",
    "rainy 🌧",
    "windy 🌬",
    "stormy 🌩"
]
```

```python
adjs = [
    'happy',
    'excited',
    'gloomy',
    'lovely',
    'sad',
    'huge',
    'lonely',
    'crazy',
    'mad',
    'perfect',
    'gentle',
    'lame',
    'precious',
    'almost crying',
    'sweet',
    'dancing',
    'cool',
    'unsatisfied',
    'furry',
    'homesick'
]
```

```python
nouns = [
    'dog',
    'car',
    'tree',
    'appetite',
    'river',
    'fire',
    'man',
    'woman',
    'leaf',
    'red house',
    'radio',
    'wooden box',
    'poster',
    'door frame',
    'painting',
    'sculpture',
    'couple',
    'globe',
    'gold',
    'train',
    'screen',
    'cat',
    'rabbit',
    'shelf',
    'book',
    'dinosaur',
    'teddybear',
    'cloud',
    'hole',
    'treasure',
    'spade',
    'cup',
    'sadness',
    'child',
    'lust',
    'balloon'
]
```

```python
advs = [
    'alone',
    'carelessly',
    'anxiously',
    'beautifully',
    'burningly',
    'powerfully',
    'curiously',
    'slowly',
    'sadly',
    'suddenly',
    'softly',
    'impatiently',
    'quickly',
    'unhappily',
    'passionately',
    'seductively',
    'tenderly',
    'wistfully'
]
```

```python
verbs = [
    'came across with',
    'saw',
    'heard',
    'jumped into',
    'kissed with',
    'hugged',
    'felt afraid of',
    'lost',
    'fell in love with',
    'fought with',
    'looked for',
    'sent',
    'sighs for',
    'tried to',
    'looked forward for',
    'completed',
    'yearns for',
    'made friend with',
    'wrote',
    'read',
    'pushed',
    'pulled',
    'got pressure from',
    'woos'
]
```

```python
# textwrap library used to "wrap" the text at a particular length
import textwrap
import random

# output begins with date and weather
output = random.choice(date) + " " + random.choice(weather) +"\n"
output += "Dear Diary"+",\n"
output += "\n"

# inside this loop, build the phrases. strachey implemented "short" phrases
# and "long" phrases; two or more "short" phrases in a row have special
# formatting rules, so we need to know what the last phrase kind was in
# order to generate the output.
history = []
body = ""
for i in range(10):
    kind = random.choice(["short", "long"])
    if kind == "long":
        # adjectives and adverbs will be present only 50% of the time
        line = " ".join([
            "Today I",
            random.choice([random.choice(advs), ""]),
            random.choice(verbs),
             "a",
            random.choice([random.choice(adjs), ""]),
            random.choice(nouns)])
        body += line
    else:
        adj_noun = random.choice(adjs) + " " + random.choice(nouns)
        # if the last phrase was "short," use truncated form
        if len(history) > 0 and history[-1] == "short":
            body += " :That was a " + random.choice(adjs) +" "+ random.choice(nouns)
        else:
            body += "I felt " + random.choice(adjs) +" like a "+ random.choice(adjs) +" "+ random.choice(nouns)
    body += ". "
    history.append(kind)
# clean up output
body = body.replace("  ", " ")
body = body.replace(". :", ":")
# put everything together
output += textwrap.fill(body, 60)

print(output)
```
