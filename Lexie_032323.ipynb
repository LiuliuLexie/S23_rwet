```python
text = open("story_of_your_life_reviewed.txt").read() # text
doc = nlp(text)
sentences = list(doc.sents)

import random
random.choice(sentences) # test
one_sentence = random.choice(sentences)
one_sentence.text # one_sentence: a string
```

```python
words = [w for w in list(doc) if w.is_alpha]
entitles = list(doc.ents)

# for i in range(10):
#     print(random.choice(words))
    
# for i in range(10):
#     print(random.choice(entitles))
```

```python
# create the word lib i need
nouns = [w for w in words if w.pos_=="NOUN"]
verbs = [w for w in words if w.pos_=="VERB"]
adjs = [w for w in words if w.pos_=="ADJ"]
advs = [w for w in words if w.pos_=="ADV"]
people = [w for w in entitles if w.label_=="PERSON"]
date = [w for w in entitles if w.label_=="DATE"]
```
